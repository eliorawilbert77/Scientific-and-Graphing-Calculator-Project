```python
#importing the functions needed
import math
import numpy as np
import matplotlib.pyplot as plt
#defining my calculator function to use
def calculator():
    print("Welcome!")
    choose = input("Do you want Calculator or Grapher?: ").lower() #.lower is so that no matter if they type uppercase or lowercase the code will work
#if the user chooses the calculator their options are displayed
    if choose == "calculator":
        print("Choose:")
        print("1. Addition")
        print("2. Subtraction")
        print("3. Division")
        print("4. Multiplication")
        print("5. Power")
        print("6. Sin")
        print("7. Cosine")
        print("8. Tangent")
        print("9. Logarithm")
        print("10. Exponential")
#for user to choose
        choice = int(input("Enter your choice (1-10): "))
#fir the first 5 calculator options you need to enter 2 different numbers and this function makes it easier to just ask for the user input once
        if choice in range(1, 11):
            if choice in [1, 2, 3, 4, 5]:
                num1 = float(input("Enter first number: "))
                num2 = float(input("Enter second number: "))
#function that does the calculations depending on what options you choose
            if choice == 1:
                result = num1 + num2
                print("Result: ", result)
            elif choice == 2:
                result = num1 - num2
                print("Result: ", result)
            elif choice == 3:
                if num2 == 0:
                    print("Error! Division by zero.")
                else:
                    result = num1 / num2
                    print("Result: ", result)
            elif choice == 4:
                result = num1 * num2
                print("Result: ", result)
            elif choice == 5:
                result = math.pow(num1, num2)
                print("Result: ", result)
            elif choice == 6:
                angle = float(input("Enter angle in degrees: "))
                result = math.sin(math.radians(angle))
                print("Result: ", result)
            elif choice == 7:
                angle = float(input("Enter angle in degrees: "))
                result = math.cos(math.radians(angle))
                print("Result: ", result)
            elif choice == 8:
                angle = float(input("Enter angle in degrees: "))
                result = math.tan(math.radians(angle))
                print("Result: ", result)
            elif choice == 9:
                num = float(input("Enter a number: "))
                if num <= 0:
                    print("Error! Logarithm undefined for non-positive numbers.")
                else:
                    result = math.log(num)
                    print("Result: ", result)
            elif choice == 10:
                num = float(input("Enter a number: "))
                result = math.exp(num)
                print("Result: ", result)
        else:
            print("Invalid choice! Please choose a number between 1 and 10.") #if the user didnt input an option given it tells them to do something in the options
    elif choose == "grapher": #if u choose the graph these the functions that are going to enact
        print("Do you have any Linear line?") #asking the user if they want to put a line with the polynomial
        linear_choose = input("Enter Yes or No: ").lower()
        if linear_choose == "yes":
          def plot_functions():
            print("Please input the linear function in the form 'mx + b'") #asking the user to do it in the y=mx+b form
    linear_function = input("Linear Function: ")
    print("Please input the polynomial function in the form 'ax^2 + bx + c'")
    polynomial_function = input("Polynomial Function: ")

    # Generate x values
    x = np.linspace(-10, 10, 400) #linespace

    # Evaluate the user-input functions for the given x values
    y1 = eval(linear_function.replace("x", "*x"))
    y2 = eval(polynomial_function.replace("x", "*x"))

    # Plot the functions
    plt.plot(x, y1, label=f'Linear Function: {linear_function}')
    plt.plot(x, y2, label=f'Polynomial Function: {polynomial_function}')

    # Add labels, title, legend, and grid
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear and Polynomial Functions')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_functions()

if linear_choose == "no":
  if linear_choose == "yes":
            # Prompt user for linear function coefficients
            m = float(input("Enter the slope (m) of the linear function: "))
            b = float(input("Enter the y-intercept (b) of the linear function: "))

            # get x values
            x = np.linspace(-10, 10, 400)

            # calculate y values for the linear function
            y1 = m * x + b

            # Plot the linear function
            plt.plot(x, y1, label=f'Linear Function: {m}x + {b}')
elif choose == "grapher":
          print("Do you want to plot a linear function?")
linear_choose = input("Enter Yes or No: ").lower()

if linear_choose == "yes":
            # Prompt user for linear function coefficients
            m = float(input("Enter the slope (m) of the linear function: "))
            b = float(input("Enter the y-intercept (b) of the linear function: "))

            # Generate x values
            x = np.linspace(-10, 10, 400)

            # Calculate y values for the linear function
            y1 = m * x + b

            # Plot the linear function
            plt.plot(x, y1, label=f'Linear Function: {m}x + {b}')
elif linear_choose == "no":
            print("Plotting only the polynomial function.")
            print("Do you want to plot a polynomial function?")
            polynomial_choose = input("Enter Yes or No: ").lower()
            if polynomial_choose == "yes":
                # Prompt user for polynomial function coefficients
                degree = int(input("Enter the degree of the polynomial function: "))
                coefficients = []
                for i in range(degree, -1, -1):
                    coeff = float(input(f"Enter the coefficient of x^{i}: "))
                    coefficients.append(coeff)

                # Generate x values
                x = np.linspace(-10, 10, 400)

                # Generate y values for the polynomial function
                y2 = np.polyval(coefficients, x)

                # Plot the polynomial function
                plt.plot(x, y2, label='Polynomial Function')
            else:
                print("Invalid choice! Please enter Yes or No.")
            else:
            print("Invalid option! Please choose Calculator or Grapher.")

    # Add labels and legend
            plt.xlabel('x')
            plt.ylabel('y')
            plt.title('Graphs')
            plt.legend()
            plt.grid(True)
            plt.show()

#used chatgpt to help with the intersection function and what def function i should use and it helped me with finding the intersection points
def find_intersection():
    print("Welcome to the point of intersection finder!")
    print("Please input the linear function in the form 'mx + b'")
    linear_function = input("Linear Function: ")
    print("Please input the quadratic function in the form 'ax^2 + bx + c'")
    quadratic_function = input("Quadratic Function: ")

    # Generate x values
    x = np.linspace(-10, 10, 400)

    # Evaluate the user-input functions for the given x values
    y_linear = eval(linear_function.replace("x", "*x"))
    y_quadratic = eval(quadratic_function.replace("x", "*x"))

    # Plot the functions
    plt.plot(x, y_linear, label=f'Linear Function: {linear_function}')
    plt.plot(x, y_quadratic, label=f'Quadratic Function: {quadratic_function}')

    # Find the intersection points
    idx = np.argwhere(np.diff(np.sign(y_linear - y_quadratic))).flatten()
    intersection_points = [(x[i], y_linear[i]) for i in idx]

    if len(intersection_points) == 0:
        print("No intersection points found.")
    else:
        print("Intersection point(s):")
        for point in intersection_points:
            print(f"({point[0]}, {point[1]})")

        # Highlight intersection points on the plot
        for point in intersection_points:
            plt.plot(point[0], point[1], 'ro')  # Mark intersection points with red dots

    # Add labels, title, legend, and grid
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Linear and Quadratic Functions')
    plt.legend()
    plt.grid(True)
    plt.show()

find_intersection()

#chatgpt helped me with the trig functions and i learned how to import them and calculate them for the graph

def plot_trigonometric_function():
    print("Welcome to the trigonometric function plotter!")
    trig_function = input("Which trigonometric function do you want to plot? (sin, cos, tan): ").lower()

    if trig_function not in ['sin', 'cos', 'tan']:
        print("Invalid input! Please choose 'sin', 'cos', or 'tan'.")
        return

    print("Please input the amplitude, frequency, and phase (in radians) of the function.")
    amplitude = float(input("Amplitude: "))
    frequency = float(input("Frequency (cycles per unit x): "))
    phase = float(input("Phase (in radians): "))

    # Generate x values
    x = np.linspace(-2*np.pi, 2*np.pi, 400)

    # Calculate y values based on the selected trigonometric function
    if trig_function == 'sin':
        y = amplitude * np.sin(frequency * x + phase)
        function_label = f'Sine Function: {amplitude}sin({frequency}x + {phase})'
    elif trig_function == 'cos':
        y = amplitude * np.cos(frequency * x + phase)
        function_label = f'Cosine Function: {amplitude}cos({frequency}x + {phase})'
    elif trig_function == 'tan':
        y = amplitude * np.tan(frequency * x + phase)
        function_label = f'Tangent Function: {amplitude}tan({frequency}x + {phase})'

    # Plot the function
    plt.plot(x, y, label=function_label)

    # Add labels, title, legend, and grid
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title(f'{trig_function.capitalize()} Function')
    plt.legend()
    plt.grid(True)
    plt.show()

plot_trigonometric_function()


calculator()
```
